```python
holidays = pd.read_csv('US Holiday Dates (2004-2021).csv')
```

```python
holidays = holidays[holidays['Year']==2018]
```

```python
holidays.reset_index(drop=True)
```

```python
holidays['Date'] = pd.to_datetime(holidays['Date'], format='%Y-%m-%d')
holidays['Date']=holidays['Date'].dt.strftime('%m/%d/%Y')
```

```python
holidays=holidays[holidays['Date']<'05/10/2018']
holidays=holidays.sort_values(by=['Month','Day'])
holidays
```

```python
holidays.to_csv('holidays.csv',index=False)
```
